```python
import tensorflow as tf
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp_df = pd.read_csv('/content/drive/MyDrive/2. Project #1/Celsius+to+Fahrenheit.csv')
```

```python
sns.scatterplot(x=temp_df['Celsius'], y=temp_df['Fahrenheit'])
```

```python
X_train = temp_df['Celsius']
y_train = temp_df['Fahrenheit']
```

```python
X_train.shape, y_train.shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1, input_shape=[1]))
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(1.0), loss='mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 500)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress during Training')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.legend(['Training Loss'])
```

```python
model.get_weights()
```

```python
Temp_C = [-32, 0, 100]
Temp_C = np.array(Temp_C)  #convert list to a numpy array
Temp_F = model.predict([Temp_C])
print(f'Temperature in DegFaren using Trained ANN = {Temp_F}')
```
